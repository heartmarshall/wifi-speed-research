```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def plot_speed_data(csv_file_path, label="Label"):
    df = pd.read_csv(csv_file_path)
    timesteps = df['Timestep']
    download_speeds = df['DownloadSpeed']
    upload_speeds = df['UploadSpeed']

    plt.figure(figsize=(10, 6))
    plt.plot(timesteps, download_speeds, label='Download Speed', linestyle='-', marker='')
    plt.plot(timesteps, upload_speeds, label='Upload Speed', linestyle='-', marker='')
    plt.title(label)
    plt.xlabel('Timestep')
    plt.ylabel('speed (Mbit/s)')

    plt.legend()
    plt.show()

def plot_boxplot_speed_data(csv_file_path, label="Label", speed_type="DownloadSpeed"):
    df = pd.read_csv(csv_file_path)

    speed_data = df[speed_type]
    mean_speed = speed_data.mean()
    max_speed = speed_data.max()


    plt.figure(figsize=(8, 6))
    sns.boxplot(y=speed_data)
    plt.title(label)

    ax = sns.boxplot(y=speed_data)
    ax.text(0.98, 0.02, f'Max: {max_speed:.2f} Mbit/s', transform=ax.transAxes, color='green', fontsize=10, ha='right')
    ax.text(0.98, 0.1, f'Mean: {mean_speed:.2f} Mbit/s', transform=ax.transAxes, color='Purple', fontsize=10, ha='right')
    plt.ylabel('Download speed (Mbit/s)')

    plt.show()
```

```python
csv_file_path = 'CSV_FILE_NAME'
plot_speed_data(csv_file_path, "LABEL")
```

```python
csv_file_path = 'CSV_FILE_NAME'
plot_boxplot_speed_data(csv_file_path, "LABEL", "UploadSpeed")
```
